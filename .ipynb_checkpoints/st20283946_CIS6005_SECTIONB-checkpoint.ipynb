##### Importing nesseccary modules

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

#####  Loading CIFAR-10 dataset

```python
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

#### Exploring the shape of data

```python
print("Training images shape:", train_images.shape)
print("Training labels shape:", train_labels.shape)
print("Test images shape:", test_images.shape)
print("Test labels shape:", test_labels.shape)
```

```python
# Explore and display images from all classes
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Display one image per class
plt.figure(figsize=(5, 3))
for i in range(10):
    class_images = train_images[train_labels.flatten() == i]
    plt.subplot(2, 5, i + 1)
    plt.imshow(class_images[0])
    plt.title(class_names[i])
    plt.axis('off')
plt.show()
```

#####  Split the training set into training and validation sets

```python
train_images, val_images, train_labels, val_labels = train_test_split(
train_images, train_labels, test_size=0.2, random_state=42)
```

#### Metadata details

```python
# Explore the shape of the data
print("Training images shape:", train_images.shape)
print("Training labels shape:", train_labels.shape)
print("Validation images shape:", val_images.shape)
print("Validation labels shape:", val_labels.shape)
print("Test images shape:", test_images.shape)
print("Test labels shape:", test_labels.shape)
```

train_images, val_images, test_images = train_images / 255.0, val_images / 255.0, test_images / 255.0

# FIRST MODEL

```python
# Define Model 1: Convolutional Neural Network (CNN)
model_cnn = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
# Compile Model 1
model_cnn.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
```

```python
model_cnn.summary()
```

```python
history_cnn = model_cnn.fit(train_images, train_labels, epochs=10, validation_data=(val_images, val_labels))
```

```python
# Print history keys
print(history_cnn.history.keys())
```

```python
# Convert history to DataFrame
metrics1 = pd.DataFrame(history_cnn.history)
metrics1
```

```python
# Plot training and validation accuracy over epochs
plt.plot(history_cnn.history['accuracy'], label='Training Accuracy')
plt.plot(history_cnn.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Plot training and validation accuracy over epochs
plt.plot(history_cnn.history['loss'], label='Training Loss')
plt.plot(history_cnn.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report

# Assuming you have trained model_cnn and have test_images, test_labels

# Make predictions using the trained model
predictions_cnn = model_cnn.predict(test_images)
predicted_labels_cnn = np.argmax(predictions_cnn, axis=1)

# Generate the classification report
class_report_cnn = classification_report(test_labels, predicted_labels_cnn)

# Display the classification report
print("Classification Report - Model 1 (CNN):\n", class_report_cnn)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Predictions on the test set
predictions = model_cnn.predict(test_images)
y_pred = np.argmax(predictions, axis=1)

# Confusion Matrix
cm = confusion_matrix(test_labels, y_pred)
plt.figure(figsize=(4, 2))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(cm)
```

```python
def visualize_prediction(model, image, actual_label, class_names):
    # Reshape the image to (1, height, width, channels) for prediction
    image = np.expand_dims(image, axis=0)

    # Perform prediction
    predictions = model.predict(image)
    
    # Get the predicted class label
    predicted_label = np.argmax(predictions)

    # Display the image
    plt.imshow(image[0])
    
    # Convert actual_label to a scalar if it's an array
    actual_label = np.squeeze(actual_label)
    
    # Display actual and predicted labels
    plt.title(f'Actual: {class_names[actual_label]}, Predicted: {class_names[predicted_label]}')
    plt.axis('off')
    plt.show()

# Assuming class_names is defined as follows
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Assuming test_images and test_labels are loaded from the CIFAR-10 test set
index = 4000# Change this to the index of the image you want to visualize
image_to_visualize = test_images[index]
actual_label_to_visualize = test_labels[index]

# Visualize the prediction for the selected image
visualize_prediction(model_cnn, image_to_visualize, actual_label_to_visualize, class_names)
```

# SECOND MODEL

```python
# Define Model 2: Convolutional Neural Network (CNN) with Different Architecture
model_cnn_2 = models.Sequential([
    layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(256, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
# Compile Model 2
model_cnn_2.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])
```

```python
model_cnn_2.summary()
```

```python
# Train Model 2 (CNN) and capture history
history_cnn_2 = model_cnn_2.fit(train_images, train_labels, epochs=10, validation_data=(val_images, val_labels))
```

```python
# Convert history to DataFrame
metrics_cnn_2 = pd.DataFrame(history_cnn_2.history)
metrics_cnn_2
```

```python
# Plot training and validation accuracy over epochs
plt.plot(history_cnn_2.history['accuracy'], label='Training Accuracy')
plt.plot(history_cnn_2.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(history_cnn_2.history['loss'], label='Training Loss')
plt.plot(history_cnn_2.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Make predictions using the trained model
predictions_cnn_2 = model_cnn_2.predict(test_images)
predicted_labels_cnn_2 = np.argmax(predictions_cnn_2, axis=1)

# Create the confusion matrix
cm_cnn_2 = confusion_matrix(test_labels, predicted_labels_cnn_2)

# Define class names (assuming you have defined class_names for CIFAR-10)
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Plot the confusion matrix
plt.figure(figsize=(8, 4))
sns.heatmap(cm_cnn_2, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Model 2 (CNN)')
plt.show()
```
